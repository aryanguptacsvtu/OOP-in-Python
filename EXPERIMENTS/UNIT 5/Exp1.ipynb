```python
import pandas as pd
import numpy as np
```

```python
# Importing CSV file into DataFrame
df = pd.read_csv('Demo.csv')

# Displaying the DataFrame
print("Original DataFrame:")
print(df)
```

```python
# Detecting missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Filling missing values with mean
df['Age'].fillna(df['Age'].mean(),inplace=True)
print("\nDataFrame after filling missing values:")
print(df)

# Dropping rows with any missing values
df_cleaned = df.dropna()
print("\nDataFrame after dropping missing values:")
print(df_cleaned)
```

```python
df['Salary(in K)'] = df['Salary'].apply(lambda df:df/1000)
print("Dataframe after apply() method:\n",df)
```

```python
df['City_uppercase'] = df['City'].map(lambda x: x.upper() if pd.notnull(x) else x)
print("\nDataFrame after transformation:")
print(df)
```

```python
# Filtering outliers in Salary column using z-score
mean = df['Salary'].mean()
std_dev = df['Salary'].std()
df_filtered = df[np.abs((df['Salary'] - mean) / std_dev) < 3]

print("\nDataFrame after filtering outliers in Salary:")
print(df_filtered)
```

```python
# Performing vectorized string operation to capitalize City names
df['City_Lowercase'] = df['City'].str.lower()

print("\nDataFrame after vectorized string operations:")
print(df)
```

```python
df.plot(x='Name', y='Age', kind='line', title='Line Plot')
```
