```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('path_to_image.jpg')
image_np = np.array(image)
```

```python
plt.imshow(image_np)
plt.title('Original image')
plt.axis('off')
plt.show()
```

```python
left = 50
right = 250
upper = 50
lower = 250

cropped_image_np = image_np[upper:lower,left:right]
flipped_image_np = np.flip(cropped_image_np, axis=1)
```

```python
plt.imshow(flipped_image_np)
plt.title('Cropped and Flipped Image')
plt.axis('off')
plt.show()
```
